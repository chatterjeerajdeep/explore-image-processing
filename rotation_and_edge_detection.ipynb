```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("/home/rajdeep/Downloads/pencil_sketch.jpeg")
```

```python
plt.imshow(img)
plt.show()
```

### Let us rotate this image

```python
rotated = cv2.rotate(img, cv2.ROTATE_90_COUNTERCLOCKWISE)
```

```python
plt.imshow(rotated)
plt.show()
```

### Let's try to perform edge detection on this image

```python
grayscale = cv2.cvtColor(rotated, cv2.COLOR_BGR2GRAY)
```

```python
blurred = cv2.GaussianBlur(grayscale, (9,9), 0)
edges = cv2.Canny(blurred, 10, 30)
```

```python
plt.imshow(edges, cmap="gray")
plt.show()
```
