### Let us rotate this image

```python
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
h,w = gray_img.shape

(cX,cY) = (w/2,h/2)
M = cv2.getRotationMatrix2D((cX,cY), 90, 1)

cos = np.abs(M[0, 0])
sin = np.abs(M[0, 1])

nW = int((h * sin) + (w * cos))
nH = int((h * cos) + (w * sin))

M[0, 2] += (nW / 2) - cX
M[1, 2] += (nH / 2) - cY
```

```python
affine_transform = cv2.warpAffine(gray_img, M, (img.shape[0], img.shape[1]))
```

```python
cv2.imshow("adjusted affine rotation", affine_transform)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
rotated = cv2.rotate(img, cv2.ROTATE_90_COUNTERCLOCKWISE)
```

### Let's try to perform edge detection on this image

```python
grayscale = cv2.cvtColor(rotated, cv2.COLOR_BGR2GRAY)
```

```python
blurred = cv2.GaussianBlur(grayscale, (9,9), 0)
edges = cv2.Canny(blurred, 10, 30)
```
